```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
with open('loss.txt') as f:
    contents = f.readlines()
Acc, Loss = [], []
for line in contents[1::2]:
    loss = float(line.split('\t')[1].split()[-1])
    acc = float(line.split('\t')[2].split()[-1])
    Loss.append(loss)
    Acc.append(acc)
```

```python
epochs = np.arange(1, 101)

plt.rcParams.update({'font.size': 16})

fig, axes = plt.subplots(1, 2, figsize=(15, 6))
axes[0].plot(epochs, Loss)
axes[0].set_title('     ')
axes[0].set_xlabel(' ')
axes[0].set_ylabel('  ')
axes[0].grid()

axes[1].plot(epochs, Acc)
axes[1].set_title(' accuracy   ')
axes[1].set_ylabel('Accuracy')
axes[1].set_xlabel(' ')
axes[1].grid()
plt.tight_layout()
plt.savefig('train.png', dpi=500)
```
