```python
import json
import numpy as np

with open('../../json/probabilities.json') as f:
    contents = json.load(f)

vals = list(contents.values())
check = [
    True if np.argmax(v[0]) == v[1] else False 
    for v in vals
]
for i, val in enumerate(check):
    if not val:
        print(i, end=' ')
```

```python
import matplotlib.pyplot as plt

X = np.array([
    np.max(v[0])
    for v in vals
])

N, bins, patches = plt.hist(X, bins=25)
print(N, len(N))
print(bins, len(bins))
plt.show()
```
